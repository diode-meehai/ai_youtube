```python
def MR_create_Xb(X):
    N = X.shape[0]
    ones = np.ones([N, 1])
    Xb = np.hstack([ones, X])
    return Xb
```

```python
def MR_find_Yhat(X, W):
    Xb = MR_create_Xb(X)
    Yhat = np.dot(Xb, W)
    return Yhat
```

```python
def find_error(Y, Yhat, TypeOfError):
    if TypeOfError == 'SSE':
        error = find_SSE(Y, Yhat)
    elif TypeOfError == 'MSE':
        error = find_MSE(Y, Yhat)
    elif TypeOfError == 'MAE':
        error = find_MAE(Y, Yhat)
    elif TypeOfError == 'MAPE':
        error = find_MAPE(Y, Yhat)
    return error
```

```python
def find_SSE(Y, Yhat):
    SSE = ((Y - Yhat)**2).sum()
    return SSE
```

```python
def find_MSE(Y, Yhat):
    N = Y.shape[0]
    SSE = ((Y - Yhat)**2).sum()
    MSE = SSE/N
    return MSE
```

```python
def find_MAE(Y, Yhat):
    N = Y.shape[0]
    MAE = (np.abs(Y - Yhat)).sum()/N
    return MAE
```

```python
def find_MAPE(Y, Yhat):
    N = Y.shape[0]
    MAPE = np.abs((Y - Yhat)/Y).sum()*100/N
    return MAPE
```

```python
def MR_find_W_local(X, Y, epoch = 1000, lr = 0.01):
    Xb = MR_create_Xb(X)
    N = Xb.shape[0]
    D_1 = Xb.shape[1] #colums
    W = np.random.randn(D_1, 1)
    error_list = []
    for i in range(epoch):
        Yhat = MR_find_Yhat(X, W)
        error = find_error(Y, Yhat, 'SSE')
        error_list.append(error)
        S = np.dot(Xb.T, Y - Yhat)
        W = W + (lr/N)*S
        #print(i)
        #print(W)
    return W, error_list
```
