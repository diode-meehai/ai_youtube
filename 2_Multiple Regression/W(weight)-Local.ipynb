```python
def MR_create_Xb(X):
    N = X.shape[0]
    ones = np.ones([N, 1])
    Xb = np.hstack([ones, X])
    return Xb
```

```python
def MR_find_W_local(X, Y, epoch = 1000, lr = 0.1):
    Xb = MR_create_Xb(X)
    N = Xb.shape[0]
    #D_1 = Xb.shape[1]
    #W = np.random.randn(D_1, 1)
    W = np.array([[0.3],[0.4],[0.5]])
    print(0)
    print(W)
    print()
    error_list = []
    for i in range(epoch):
        print(i+1)
        #Yhat = MR_find_Yhat(X, W)
        #error = find_error(Y, Yhat, 'SSE')
        #error_list.append(error)
        #S = np.dot(Xb.T, Y - Yhat)
        #W = W + (lr/N)*S
        Yhat = np.dot(Xb, W)
        W = W + (lr/N)*np.dot(Xb.T, Y-Yhat)
        print(W)
        print()
    return W, error_list
```
