```python
tf.compat.v1.disable_eager_execution()
#ประกาศตัวแปร
#X = tf.placeholder(tf.float32, [None, 2], name='X') #ที่ยังไม่ได้ถูกกำหนดค่า tf.placeholder(type)
X = tf.compat.v1.placeholder(tf.float32, [None, 2], name='X')
#Y = tf.placeholder(tf.float32, [None, 1], name='Y')
Y = tf.compat.v1.placeholder(tf.float32, [None, 1], name='Y')

thetas = tf.Variable(tf.zeros([2, 1]), name='Thetas') #tf.Variable ตัวแปรมีการเปลี่ยนแปลงตลอดเวลา
theta0 = tf.Variable(tf.zeros([1]), name='theta0')
```

```python
theta0
```

```python
'''
logistic (simoid)
hθ(x) = 1/(1 + e^(−θ.T x))
g(z) = 1/(1 + e^(−z))
'''
z = tf.matmul(X, thetas) # X1,X2,...Xn
#hypothesis_function = tf.div(1.0,1.0 + tf.exp(-(z + theta0)))  # theta0 คือ θ ศูนย์
hypothesis_function = tf.compat.v1.div(1.0,1.0 + tf.exp(-(z + theta0)))  # theta0 คือ θ ศูนย์
```

```python
'''
Step 3: Define the cost function
-1/m[Σ y^(i) log(hθ(x^(i))) + (1 − y^(i))log(1 − hθ(x^(i)))]
'''

#cost_function = tf.reduce_mean(-(Y * tf.log(hypothesis_function)) - (1 - Y) * tf.log(1 - hypothesis_function)) # write the cost function based on TF methods here
cost_function = tf.compat.v1.reduce_mean(-(Y * tf.compat.v1.log(hypothesis_function)) - (1 - Y) * tf.compat.v1.log(1 - hypothesis_function)) # write the cost function based on TF methods here
```

```python
# θj := θj − α∂/∂θj * J(θ0, θ1)
#optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.03).minimize(cost_function)
optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.03).minimize(cost_function)
```

```python
with tf.compat.v1.Session() as session:
  
    # write your code for initializing all variables
    session.run(tf.compat.v1.global_variables_initializer())

    for i in range(30000):
        #Run 2 ตัวแปร เลยต้องใส่ optimizer กับ cost_function => tuple python
        #https://web.facebook.com/meetdrtanet/posts/1486387858098663/?_rdc=1&_rdr
        _, cost = session.run([optimizer, cost_function], feed_dict={X: Input, Y: Output})# run the optimizer and cost function in a session w.r.t. matrices Input and Output})

        print("Epoch: {0}, cost = {1}".format(i + 1, cost))

    print("theta0: {0}, theta1: {1}, theta2: {2}".format(session.run(theta0[0]), session.run(thetas[0][0]), session.run(thetas[1][0])))
```
