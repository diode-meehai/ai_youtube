```python
import nbimporter
import myai as ai

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Data = pd.read_excel('ANN Price prediction.xlsx', usecols = 'E:O', skiprows = range(1, 5))
```

```python
Data
```

### 1.1 change dataframe to matrix

```python
DataMatrix = Data.values[:-1, :]
```

```python
np.random.shuffle(DataMatrix)
```

### 1.2 split matrix to X and Y

```python
N = DataMatrix.shape[0]
```

```python
D = DataMatrix.shape[1] - 1
```

```python
X = DataMatrix[:, :D]
Y = DataMatrix[:, D:]
```

### 1.3 split data to training set and test set

```python
start = 0
end = int(0.8*N)
```

```python
X_train = np.array(X[start:end, :], dtype='float32')
Y_train = np.array(Y[start:end, :], dtype='float32')

X_test = np.array(X[end:, :], dtype='float32')
Y_test = np.array(Y[end:, :], dtype='float32')
```

### 1.4 normalize X and Y

```python
X_train
```

```python
norm_X = ai.Normalize(X_train, norm_type='minmaxNorm')
norm_Y = ai.Normalize(Y_train, norm_type='minmaxNorm')
```

```python
norm_X.Data_min
```

```python
X_train_norm = norm_X.fit(X_train)
Y_train_norm = norm_Y.fit(Y_train)
```

```python
X_test_norm = norm_X.fit(X_test)
Y_test_norm = norm_Y.fit(Y_test)
```

## 2. Create Model

### 2.1 learn

```python
regr = ai.NN([10, 1], ['sigmoid', 'relu'])
```

```python
regr.fit(X_train_norm, Y_train_norm, loss_function='mse', epoch=100, learning_rate=0.1)
```

```python
plt.plot(regr.loss_list)
```

### 2.2 model evaluation for traning set

```python
Yhat_train_norm = regr.predict(X_train_norm)
```

```python
Yhat_train = norm_Y.transform(Yhat_train_norm)
```

```python
ai.ErrorScore.find_error(Y_train, Yhat_train, 'MSE')
```

```python
plt.scatter(Yhat_train, Y_train)
```

## 3. Making Prediction

### 3.1 Predict

```python
Yhat_test_norm = regr.predict(X_test_norm)
```

```python
Yhat_test = norm_Y.transform(Yhat_test_norm)
```

### 3.2 model evaluation for test set

```python
ai.ErrorScore.find_error(Y_test, Yhat_test, 'MSE')
```

```python
plt.scatter(Yhat_test, Y_test)
```
