```python
import nbimporter
import myai as ai

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Data = pd.read_csv('Quora Word2Vec.csv')
```

```python
Data
```

### 1.1 change dataframe to matrix

```python
DataMatrix = Data.values
```

```python
np.random.shuffle(DataMatrix)
```

### 1.2 split matrix to X and Y

```python
N = DataMatrix.shape[0]
```

```python
D = DataMatrix.shape[1] - 1
```

```python
X = DataMatrix[:, :D]
Y = DataMatrix[:, D:]
```

### 1.3 split data to training set and test set

```python
start = 0
end = int(0.8*N)
```

```python
X_train = np.array(X[start:end, :], dtype='float32')
Y_train = np.array(Y[start:end, :], dtype='float32')

X_test = np.array(X[end:, :], dtype='float32')
Y_test = np.array(Y[end:, :], dtype='float32')
```

### 1.4 normalize X and Y

```python
# X_train_norm = norm_X.fit(X_train)
# X_test_norm = norm_X.fit(X_test)
```

## 2. Create Model

### 2.1 learn

```python
clf = ai.RNN(n_ts_feature=60, n_ts_target=1, RNN_HL=[3], ANN_HL=[2, 1], ANN_AF=[['prelu', 0.5], 'sigmoid'])
```

```python
clf.fit(X_train, Y_train, 'm2_entropy', epoch=40, learning_rate=0.1)
```

```python
plt.plot(clf.loss_list)
```

### 2.2 model evaluation for training set

```python
Yhat_train_TS = clf.predict(X_train)
```

```python
Y_train_TS = ai.RNN.seperate_timestep(Y_train, clf.n_ts_target)
```

```python
for i in range(clf.n_ts_target):
    print('timestep :', i+1)
    print(ai.ErrorScore.find_error(Y_train_TS[i], Yhat_train_TS[i].data, 'binary'))
```

## 3. Making Prediction

### 3.1 Predict

```python
Yhat_test_TS = clf.predict(X_test)
```

### 3.2 model evaluation for test set

```python
Y_test_TS = ai.RNN.seperate_timestep(Y_test, clf.n_ts_target)
```

```python
for i in range(clf.n_ts_target):
    print('timestep :', i+1)
    print(ai.ErrorScore.find_error(Y_test_TS[i], Yhat_test_TS[i].data, 'binary'))
```
